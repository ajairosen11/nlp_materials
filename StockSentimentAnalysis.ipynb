```python
import pandas as pd
```

```python
df=pd.read_csv('Data.csv',encoding='ISO-8859-1')
```

```python
df.sample(2)
```

```python
df.shape
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
#Removing punctuations
data=train.iloc[:,2:]
data=data.replace(r'[^A-Za-z]',' ',regex=True)

#Renaming column names for ease of access
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head(2)
```

```python
# Converting headlines to lower case
data[new_index] = data[new_index].apply(lambda col: col.str.lower())
data.head(2)
```

```python
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#Implement Bag of words
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
#Implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
#Transform for testing data

data_test=test.iloc[:,2:]
data_test=data_test.replace(r'[^A-Za-z]',' ',regex=True)
list1_test=[i for i in range(25)]
new_index_test=[str(i) for i in list1_test]
data_test.columns=new_index_test
data_test[new_index_test] = data_test[new_index_test].apply(lambda col: col.str.lower())
headlines_test=[]
for row in range(0,len(data_test.index)):
    headlines_test.append(' '.join(str(x) for x in data_test.iloc[row,0:25]))

test_dataset = countvector.transform(headlines_test)
```

```python
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
score=accuracy_score(test['Label'],predictions)
print(score)
```

```python
import seaborn as sns
matrix=confusion_matrix(test['Label'],predictions)
sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues")
```
